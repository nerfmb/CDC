```python

import time
locations = ['Rome', 'Amsterdam','Paris','Tuscany']
location_data = []

for i in locations:
    response = requests.get("http://tour-pedia.org/api/getReviews?location=" + i + "&language=en")
    buffer = response.json()
    for x in range(len(buffer)):
        buffer[x]['location'] = i
    location_data = location_data + buffer
    print("Sleeping")
    if i != 'Tuscany':
        time.sleep(30)
    print("ah") 
```
